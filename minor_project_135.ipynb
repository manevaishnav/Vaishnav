```python
import csv

def read_Chemical_Eds_from_csv(file_path):
    reactions = []
    with open(file_path, 'r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip the header row
        for row in reader:
            reaction = {
                'Reactant 1': row[0],
                'Reactant 2': row[1],
                'Product': row[2],
                'Rate': float(row[3]),
                'Temperature': int(row[4])
            }
            reactions.append(reaction)
    return reactions

def find_reactions_by_temperature(reactions, temperature):
    matching_reactions = []
    for reaction in reactions:
        if reaction['Temperature'] == temperature:
            matching_reactions.append(reaction)
    return matching_reactions

def find_reactions_by_reactant(reactions, reactant):
    matching_reactions = []
    for reaction in reactions:
        if reaction['Reactant 1'] == reactant or reaction['Reactant 2'] == reactant:
            matching_reactions.append(reaction)
    return matching_reactions

def find_fastest_reaction(reactions):
    fastest_reaction = max(reactions, key=lambda r: r['Rate'])
    return fastest_reaction

def find_slowest_reaction(reactions):
    slowest_reaction = min(reactions, key=lambda r: r['Rate'])
    return slowest_reaction

def find_common_products(reactions):
    common_products = {}
    for reaction in reactions:
        product = reaction['Product']
        if product in common_products:
            common_products[product] += 1
        else:
            common_products[product] = 1
    return common_products

# Example usage:
reactions = read_reactions_from_csv('Chemical_Eds.csv')

# Find reactions at a specific temperature
temperature = 30
matching_reactions = find_reactions_by_temperature(reactions, temperature)
print(f"Reactions at temperature {temperature}°C:")
for reaction in matching_reactions:
    print(f"{reaction['Reactant 1']} + {reaction['Reactant 2']} -> {reaction['Product']}")

# Find reactions involving a specific reactant
reactant = 'H2O2'
matching_reactions = find_reactions_by_reactant(reactions, reactant)
print(f"Reactions involving {reactant}:")
for reaction in matching_reactions:
    print(f"{reaction['Reactant 1']} + {reaction['Reactant 2']} -> {reaction['Product']}")

# Find the fastest and slowest reactions
fastest_reaction = find_fastest_reaction(reactions)
print(f"The fastest reaction: {fastest_reaction['Reactant 1']} + {fastest_reaction['Reactant 2']} -> {fastest_reaction['Product']}")

slowest_reaction = find_slowest_reaction(reactions)
print(f"The slowest reaction: {slowest_reaction['Reactant 1']} + {slowest_reaction['Reactant 2']} -> {slowest_reaction['Product']}")

# Find the common products and their frequencies
common_products = find_common_products(reactions)
print("Common products and their frequencies:")
for product, frequency in common_products.items():
    print(f"{product}: {frequency}")

# Print the title of your minor project
print("Title: Chemical Reactions Analysis")

# Additional code for your project can be added here
```

```python
import csv
import matplotlib.pyplot as plt

def read_Chemical_Eds_from_csv(file_path):
    reactions = []
    with open(file_path, 'r') as file:
        reader = csv.reader(file)
        next(reader)  # Skip the header row
        for row in reader:
            reaction = {
                'Reactant 1': row[0],
                'Reactant 2': row[1],
                'Product': row[2],
                'Rate': float(row[3]),
                'Temperature': int(row[4])
            }
            reactions.append(reaction)
    return reactions

def find_reactions_by_temperature(reactions, temperature):
    matching_reactions = []
    for reaction in reactions:
        if reaction['Temperature'] == temperature:
            matching_reactions.append(reaction)
    return matching_reactions

def find_reactions_by_reactant(reactions, reactant):
    matching_reactions = []
    for reaction in reactions:
        if reaction['Reactant 1'] == reactant or reaction['Reactant 2'] == reactant:
            matching_reactions.append(reaction)
    return matching_reactions

def plot_reactions_by_temperature(reactions):
    temperatures = []
    rates = []
    for reaction in reactions:
        temperatures.append(reaction['Temperature'])
        rates.append(reaction['Rate'])

    plt.plot(temperatures, rates, 'ro')
    plt.xlabel('Temperature (°C)')
    plt.ylabel('Rate (mole/s)')
    plt.title('Reaction Rates at Different Temperatures')
    plt.show()

def generate_reactant_pie_chart(reactions):
    reactants = {}
    for reaction in reactions:
        reactant1 = reaction['Reactant 1']
        reactant2 = reaction['Reactant 2']
        reactants[reactant1] = reactants.get(reactant1, 0) + 1
        reactants[reactant2] = reactants.get(reactant2, 0) + 1

    labels = reactants.keys()
    sizes = reactants.values()

    plt.pie(sizes, labels=labels, autopct='%1.1f%%')
    plt.title('Reactant Composition in Reactions')
    plt.axis('equal')
    plt.show()

# Example usage:
reactions = read_reactions_from_csv('Chemical_Eds.csv')

# Problem Statement 1: Find reactions at a specific temperature and plot the rate
temperature = 30
matching_reactions = find_reactions_by_temperature(reactions, temperature)
print(f"Reactions at temperature {temperature}°C:")
for reaction in matching_reactions:
    print(f"{reaction['Reactant 1']} + {reaction['Reactant 2']} -> {reaction['Product']}")
plot_reactions_by_temperature(matching_reactions)

# Problem Statement 2: Find reactions involving a specific reactant and generate a pie chart
reactant = 'H2O2'
matching_reactions = find_reactions_by_reactant(reactions, reactant)
print(f"Reactions involving {reactant}:")
for reaction in matching_reactions:
    print(f"{reaction['Reactant 1']} + {reaction['Reactant 2']} -> {reaction['Product']}")
generate_reactant_pie_chart(matching_reactions)

# Problem Statement 3: Plot reaction rates at different temperatures
all_reactions = read_reactions_from_csv('Chemical_Eds.csv')
plot_reactions_by_temperature(all_reactions)

# Problem Statement 4: Generate a pie chart of reactant composition in all reactions
generate_reactant_pie_chart(all_reactions)

# Problem Statement 5: Find the reaction with the highest rate
reaction_with_highest_rate = max(reactions, key=lambda x: x['Rate'])
print("Reaction with the highest rate:")
print(f"{reaction_with_highest_rate['Reactant 1']} + {reaction_with_highest_rate['Reactant 2']} -> {reaction_with_highest_rate['Product']}")

# Problem Statement 6: Find the average rate of all reactions
total_rates = sum(reaction['Rate'] for reaction in reactions)
average_rate = total_rates / len(reactions)
print(f"Average rate of all reactions: {average_rate} mole/s")

# Problem Statement 7: Count the number of reactions at each temperature
temperature_counts = {}
for reaction in reactions:
    temperature = reaction['Temperature']
    temperature_counts[temperature] = temperature_counts.get(temperature, 0) + 1
print("Number of reactions at each temperature:")
for temperature, count in temperature_counts.items():
    print(f"{temperature}°C: {count} reactions")

# Problem Statement 8: Plot a bar graph of the number of reactions at each temperature
temperatures = list(temperature_counts.keys())
counts = list(temperature_counts.values())
plt.bar(temperatures, counts)
plt.xlabel('Temperature (°C)')
plt.ylabel('Number of Reactions')
plt.title('Number of Reactions at Different Temperatures')
plt.show()

# Problem Statement 9: Find the reactant that appears in the highest number of reactions
reactant_counts = {}
for reaction in reactions:
    reactant1 = reaction['Reactant 1']
    reactant2 = reaction['Reactant 2']
    reactant_counts[reactant1] = reactant_counts.get(reactant1, 0) + 1
    reactant_counts[reactant2] = reactant_counts.get(reactant2, 0) + 1
most_common_reactant = max(reactant_counts, key=reactant_counts.get)
print(f"The most common reactant in the reactions is: {most_common_reactant}")

# Problem Statement 10: Generate a pie chart of the most common reactant composition
most_common_reactant_reactions = find_reactions_by_reactant(reactions, most_common_reactant)
generate_reactant_pie_chart(most_common_reactant_reactions)
```
